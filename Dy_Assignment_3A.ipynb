````python
# month
import re
import csv   
reader = csv.reader(open('recweekdata2.csv','r'))

dirty_month_list = []
for row in reader: 
    dirty_month_list.append(row[2])
    clean_month_list = [re.sub(r'(January|1|Jan)',r'January',month)for month in dirty_month_list]
    clean_month_list = [re.sub(r'(February|2|Feb|Febuaryruary)',r'February',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(March|3|Mar)',r'March',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(April|4|Apr)',r'April',month)for month in clean_month_list] 
    clean_month_list = [re.sub(r'(May|5|May)',r'May',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(June|6|Jun)',r'June',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(July|7|Jul)',r'July',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(August|8|Aug)',r'August',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(September|9|Sep)',r'September',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(October|10|Oct|January0)',r'October',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(November|11|JanuaryJanuary|Nov)',r'November',month)for month in clean_month_list]
    clean_month_list = [re.sub(r'(December|12|JanuaryFebruary|Dec)',r'December',month)for month in clean_month_list]
print(clean_month_list)
````

````python
# course 
import re
import csv   
reader = csv.reader(open('recweekdata2.csv','r'))

dirty_course_list = []
for row in reader: 
    dirty_course_list.append(row[5])
    clean_course_list = [re.sub(r'(BS LM|Legal Management|LM)',r'Legal Management',course)for course in dirty_course_list]
    clean_course_list = [re.sub(r'(BS MGT|Management|MGT)',r'Management',course)for course in clean_course_list]
    clean_course_list = [re.sub(r'(BS CTM|Communications Technology|CTM)',r'Communications Technology',course)for course in clean_course_list]
    clean_course_list = [re.sub(r'(BS ITE|Information Technology Entrepreneurship|ITE)',r'Information Technology Entrepreneurship',course)for course in clean_course_list]
    clean_course_list = [re.sub(r'(BS ME|Management Engineering|ME|Management Management Engineering)',r'Management Engineering',course)for course in clean_course_list]

print(clean_course_list)
````
